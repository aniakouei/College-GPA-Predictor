```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("sat_gpa_v1.csv")

data.head()
```

```python
sns.set()
sns.lmplot(x='sat_sum', y='fy_gpa', data=data)
plt.title('Relationship between GPA and SAT Scores')
plt.xlabel('SAT Scores')
plt.ylabel('GPA')
plt.show()
```

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(data['sat_sum'])
y = data['fy_gpa']


model = sm.OLS(y, X).fit()


print(model.summary())
```
